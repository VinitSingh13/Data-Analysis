```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Vinit\Downloads\movies.csv')
```

```python
import numpy as np
```

```python
for cols in df.columns:
    pct_missing = np.mean(df[cols].isnull())
    print('{} {}%'.format(cols,pct_missing))
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int')
df['gross'] = df['gross'].astype('int')
df.head()
```

```python
df['released year'] = df['released'].str[0:4]
df.head()
```

```python
df = df.sort_values(by='gross',inplace=False,ascending=False)
```

```python
df.head()
```

```python
pd.set_option('display.max_row',None)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
plt.scatter(df['budget'],df['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
```

```python
import seaborn as sns
sns.regplot(x=df['budget'],y=df['gross'],data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')
```

```python
object_to_digits = df
for cols in object_to_digits.columns:
    if object_to_digits[cols].dtype == 'object':
        object_to_digits[cols] = object_to_digits[cols].astype('category')
        object_to_digits[cols] = object_to_digits[cols].cat.codes
object_to_digits.head()        
```

```python
fig,axes = plt.subplots(figsize=(13,8))
sns.heatmap(object_to_digits.corr(),annot=True)
plt.title('Correlation for every single element in data')
```
